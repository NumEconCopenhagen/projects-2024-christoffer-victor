```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the first CSV file
df1 = pd.read_csv('nat.csv', sep=";", decimal="," )

# Load the second CSV file
df2 = pd.read_csv('Work.csv', sep=";", decimal="," )

# Select the desired columns from the first dataset
df1_selected = df1[['Year','Y', 'M', 'X', 'C', 'G', 'I']]

# Select the desired columns from the second dataset
df2_selected = df2[['N']]

# Combine the selected columns into a new dataset
df_combined = pd.concat([df1_selected, df2_selected], axis=1)
```

```python
# Create a new column by subtracting 'M' from 'X'
df_combined['NX'] = df_combined['X'] - df_combined['M']
```

```python
# Convert columns to numeric, excluding 'Year'
for column in df_combined.columns:
    if column != 'Year':
        df_combined[column] = pd.to_numeric(df_combined[column], errors='coerce')

# Create a new DataFrame for the growth rates
df_growth = pd.DataFrame()
df_growth['Year'] = df_combined['Year']

# Calculate the growth rates and add them as new columns, excluding 'Year'
for column in df_combined.columns:
    if column != 'Year':
        df_growth[column + '_growth'] = df_combined[column].pct_change() * 100  # Multiply by 100 to get percentage

# Remove the year 2000 from the DataFrame
df_growth = df_growth[df_growth['Year'] != 2000]

# Print the DataFrame with the new growth rate columns
print(df_growth)
```

```python
# Set the style of the plot
plt.style.use('ggplot')  # Use 'ggplot' style

# Create a line plot for each column in the DataFrame, excluding 'Year'
for column in df_growth.columns:
    if column != 'Year':
        plt.plot(df_growth['Year'], df_growth[column], label=column)

# Add a title and labels to the plot
plt.title('Growth Rates Over Time')
plt.xlabel('Year')
plt.ylabel('Growth Rate (%)')

# Add a legend to the plot
plt.legend()

# Display the plot
plt.show()
```
