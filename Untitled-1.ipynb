```python
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as widgets
import bqplot as bq
from IPython.display import display

# Load the first CSV file
df1 = pd.read_csv('nat.csv', sep=";", decimal="," )

# Load the second CSV file
df2 = pd.read_csv('Work.csv', sep=";", decimal="," )

# Select the desired columns from the first dataset
df1_selected = df1[['Year','Y', 'M', 'X', 'C', 'G', 'I']]

# Select the desired columns from the second dataset
df2_selected = df2[['N']]

# Combine the selected columns into a new dataset
df_combined = pd.concat([df1_selected, df2_selected], axis=1)
```

```python
# Create a new column by subtracting 'M' from 'X'
df_combined['NX'] = df_combined['X'] - df_combined['M']
```

```python
# Convert columns to numeric, excluding 'Year'
for column in df_combined.columns:
    if column != 'Year':
        df_combined[column] = pd.to_numeric(df_combined[column], errors='coerce')

# Create a new DataFrame for the growth rates
df_growth = pd.DataFrame()
df_growth['Year'] = df_combined['Year']

# Calculate the growth rates and add them as new columns, excluding 'Year'
for column in df_combined.columns:
    if column != 'Year':
        df_growth[column + '_growth'] = df_combined[column].pct_change() * 100  # Multiply by 100 to get percentage

# Remove the year 2000 from the DataFrame
df_growth = df_growth[df_growth['Year'] != 2000]

# Print the DataFrame with the new growth rate columns
print(df_growth)
```

```python
# Set the style of the plot
plt.style.use('ggplot')  # Use 'ggplot' style

# Create a bar chart for each column in the DataFrame, excluding 'Year', 'M_growth', and 'X_growth'
bar_width = 0.1
for i, column in enumerate(df_growth.columns):
    if column not in ['Year', 'M_growth', 'X_growth']:
        plt.bar(df_growth['Year'] + i * bar_width, df_growth[column], width=bar_width, label=column)

# Add a title and labels to the plot
plt.title('Growth Rates Over Time')
plt.xlabel('Year')
plt.ylabel('Growth Rate (%)')

# Add a legend to the plot
plt.legend()

# Display the plot
plt.show()
```

```python
# Create a dropdown menu with the variable names
dropdown = widgets.Dropdown(options=['Y', 'C', 'G', 'I', 'NX', 'N'])

# Create a dictionary that maps the original variable names to the new names
name_dict = {'Y': 'GDP', 'C': 'Consumption', 'G': 'Public consumption', 'I': 'Investment', 'NX': 'Net exports', 'N': 'Employment level'}

# Create scales for the x and y axes
x_sc = bq.LinearScale()
y_sc = bq.LinearScale()

# Create a line mark
line = bq.Lines(x=df_combined['Year'], y=[], scales={'x': x_sc, 'y': y_sc})

# Create axes
x_ax = bq.Axis(scale=x_sc, label='Year')
y_ax = bq.Axis(scale=y_sc, label='Value', orientation='vertical')

# Create a figure
fig = bq.Figure(marks=[line], axes=[x_ax, y_ax], title='')

# Create a function to update the figure
def update_figure(change):
    line.y = df_combined[dropdown.value]
    fig.title = f'{name_dict[dropdown.value]} development from 2000 to 2023'

# Update the figure when the dropdown value changes
dropdown.observe(update_figure, 'value')

# Display the dropdown and the figure
display(dropdown)
display(fig)

# Initialize the figure
update_figure(None)
```
